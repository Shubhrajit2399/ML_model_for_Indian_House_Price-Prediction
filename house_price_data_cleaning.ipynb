```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("house_prices.csv")
df.info()
```

```python
def convert_to_int(x):
    if pd.isna(x):
        return 0
    if isinstance(x, str):
        # Handle '> 10' case
        if '>' in x:
            try:
                # Extract the number after '>' and add 1
                return int(x.split('>')[1].strip()) + 1
            except:
                pass
        # Try to convert other string values
        try:
            return int(x)
        except:
            return 0  # Default for non-convertible strings
    return int(x)

def convert_title(y):
    if isinstance(y, str):
        if 'BHK' in y:
            try:
                bhk_parts=y.split('BHK')[0].strip().split()
                return int(bhk_parts[-1])
            except:
                return 0
        try:
            return int(y)
        except:
            return 0
    return 0 if pd.isna(y) else int(y)      

df["Title(BHK)"]=df["Title"].apply(convert_title)
df_trm1=df[["Title(BHK)","Amount(in rupees)","location","Carpet Area","Floor","Transaction","Furnishing","facing","overlooking","Bathroom","Balcony","Car Parking"]].copy()
df_trm1["Bathroom"]=df_trm1["Bathroom"].apply(convert_to_int)
df_trm1["Balcony"]=df_trm1["Balcony"].apply(convert_to_int)

df_trm1.info()
```

```python
df_trm1.head()
```

```python
def trim_carpet(z):
    if pd.isna(z):
        return 0
    if isinstance(z,str):
        if 'sqft' in z:
            try:
                ar_parts=z.split('sqft')[0].strip().split()
                return int(ar_parts[-1])
            except:
                return 0
        try:
            return int(z)
        except:
            return 0
    return 0 if pd.isna(z) else int(z)

df_trm1["Carpet Area(sqft)"]=df_trm1["Carpet Area"].apply(trim_carpet)
df_trm1.info()
```

```python
df_trm1.head()
```

```python
mask=df_trm1["Carpet Area(sqft)"]==0
if mask.any():
    df_trm1=df_trm1[~mask]
df_trm1=df_trm1.dropna()
df_trm1.info()
```

```python
mask=df_trm1["Amount(in rupees)"]=="Call for Price"
if mask.any():
    df_trm1=df_trm1[~mask]
df_trm1=df_trm1.dropna()
df_trm1.info()
```

```python
df_trm1[["Parking Spaces","Parking Type"]]=df_trm1["Car Parking"].str.split(" ",n=1,expand=True)
df_trm1[["Floor No.","Total Floors"]]=df_trm1["Floor"].str.split(" out of ",n=1,expand=True)
df_trm1[["Ammount","Ammount_Idx"]]=df_trm1["Amount(in rupees)"].str.split(" ",n=1,expand=True)
df_trm1.info()
```

```python
df_trm1.head()
```

```python
df_trm1.loc[df_trm1["Floor No."]=="Ground","Floor No."]="0"
```

```python
def prop_num_cnv(w):
    if pd.isna(w):
        return 0
    if isinstance(w,str):
        try:
            return int(w)
        except:
            return 0
    return 0 if isna(w) else int(w)

def prop_prc_cnv(v):
    if pd.isna(v):
        return 0
    if isinstance(v,str):
        try:
            return float(v)
        except:
            return 0
    return 0 if isna(v) else float(v)

df_trm1["Parking Spaces"]=df_trm1["Parking Spaces"].apply(prop_num_cnv)
df_trm1["Floor No."]=df_trm1["Floor No."].apply(prop_num_cnv)
df_trm1["Total Floors"]=df_trm1["Total Floors"].apply(prop_num_cnv)
df_trm1["Ammount"]=df_trm1["Ammount"].apply(prop_prc_cnv)
df_trm1.info()
```

```python
mask=df_trm1["Ammount_Idx"]=="Lac"
df_trm1.loc[mask,"Price"]=df_trm1.loc[mask,"Ammount"]*100000
mask1=df_trm1["Ammount_Idx"]=="Cr"
df_trm1.loc[mask1,"Price"]=df_trm1.loc[mask1,"Ammount"]*10000000
df_trm1.info()
```

```python
mask=df_trm1["Carpet Area(sqft)"]>10000
if mask.any():
    df_trm1=df_trm1[~mask]
df_trm1=df_trm1.dropna()
df_trm1.info()
```

```python
df_trim_prop=df_trm1[["Title(BHK)","Price","location","Carpet Area(sqft)","Floor No.","Total Floors","Transaction","Furnishing","facing","overlooking","Bathroom","Balcony","Parking Spaces","Parking Type"]]
df_trim_prop.info()
```

```python
mask=df_trim_prop["Parking Spaces"]>5
if mask.any():
    df_trim_prop=df_trim_prop[~mask]
df_trim_prop=df_trim_prop.dropna()
df_trim_prop.info()
```

```python
mask=df_trim_prop["Price"]>(1e9/4)
if mask.any():
    df_trim_prop=df_trim_prop[~mask]
df_trim_prop=df_trim_prop.dropna()
df_trim_prop.info()
```

```python
df_trim_prop.to_csv("house_price_trim2.csv",index=False)
```
