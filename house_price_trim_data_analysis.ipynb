```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("house_price_trim2.csv")
df.info()
```

```python
prop_loc_grp=df.groupby(["location","Transaction","Furnishing"])["Title(BHK)"].count().reset_index(name="Count")
print(prop_loc_grp.info())
prop_loc_grp
```

```python
plt.figure(figsize=(14,20))
plt.barh(prop_loc_grp["location"],prop_loc_grp["Count"],zorder=2)
plt.xlabel("No. of Property")
plt.ylabel("Location")
plt.grid()
plt.show()
```

```python
plt.scatter(df["Price"],df["Carpet Area(sqft)"],zorder=2)
plt.xlabel("Price")
plt.ylabel("Carpet Area")
plt.grid()
plt.show()
```
